```python


# Prepare the data
data = Diwali1.groupby('Gender')['Amount'].count()
labels = data.index  # Use Gender categories for labels


# Plot the pie chart
plt.pie(
    data, 
    labels=labels, 
    explode=(0.08, 0), 
    autopct='%0.0f%%', 
    colors = sns.color_palette('Paired')[0:4]
)
plt.suptitle("Gender vs Amount Distribution")
plt.savefig("1.png")  # Save the plot
plt.show()  # Display the plot
```

```python
Age_Gen=sns.countplot(x='Age Group',data=Diwali1,hue='Gender',palette='BrBG')
for i in Age_Gen.containers:
    Age_Gen.bar_label(i)
```

```python


# Drop duplicates based on 'Gender' and 'Occupation'
df_temp = Diwali1.drop_duplicates(subset=['Gender', 'Occupation'])

# Count the top 10 occurrences of 'Gender'
x = df_temp['Gender'].value_counts().head(10)

# Create the plot
plt.figure(figsize=(17, 8))
label = sns.countplot(
    data=df_temp, 
    x='Gender', 
    hue='Age Group', 
    order=x.index, 
    palette='summer'
)

# Add labels to the bars
for i in label.containers:
    label.bar_label(i)

# Save and display the plot
plt.savefig("2.png")  # Save before displaying
plt.show()
```

```python
# top 10 state with most Diwali orders
filtered_state=Diwali1[Diwali1['State']!='unavilable']
filtered_state
```

```python
top_state=filtered_state['State'].value_counts().head(10)
top_state
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_state.values, y=top_state.index, palette='plasma')
plt.title('Top 10 States with Most Diwali Sales (Excluding Unknown)')
plt.xlabel('Number of Orders')
plt.ylabel('State')
plt.show()
```

```python
# top 10 most sold products (same thing as above)

palette = sns.color_palette('Set2', 10)  # Using Set2 palette with 10 colors

fig1, ax1 = plt.subplots(figsize=(12, 7))
top_10_products = Diwali1.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False)

top_10_products.plot(kind='bar', color=palette, ax=ax1)

ax1.set_title('Top 10 Most Sold Products', fontsize=16)
ax1.set_xlabel('Product ID', fontsize=14)
ax1.set_ylabel('Total Orders', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
